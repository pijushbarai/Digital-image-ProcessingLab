```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
plt.imshow(rgb)
```

```python
grayscale = cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale,cmap='gray')
```

```python
kernel = np.ones([3,3])/9
kernel
```

```python
filtered_image = cv2.filter2D(grayscale,-1,kernel)
plt.imshow(filtered_image,cmap='gray')
```

```python

w,h = grayscale.shape
print(w,h)
newimage = np.zeros([w-2,h-2])
print(newimage.shape)
for i in range(w-2):
    for j in range(h-2):
        mat = grayscale[i:i+3,j:j+3]
        newimage[i,j] = np.sum(np.multiply(kernel,mat))
plt.imshow(newimage,cmap='gray')
```
