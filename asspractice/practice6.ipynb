```python

import cv2
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
print(rgb.shape)
plt.imshow(rgb)
```

```python
grayscale= cv2.cvtColor(rgb,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale,cmap='gray')
```

```python
#adding salt and pepeer noise
w,h = grayscale.shape
noisenumber = w*h//16
noisenumber
for i in range(noisenumber):
    val = random.randint(0,2)
    x = random.randint(0,w-1)
    y = random.randint(0,h-1)
    grayscale[x,y] = val*255
plt.imshow(grayscale,cmap='gray')
```

```python
newimage = cv2.GaussianBlur(grayscale,(5,5),0)
plt.imshow(newimage,cmap='gray')
```
