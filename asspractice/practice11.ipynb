```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
print(rgb.shape)
plt.imshow(rgb)
```

```python
grayscale = cv2.cvtColor(rgb,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale,cmap='gray')
```

```python
fftimg = np.fft.fft2(grayscale)
centeredfft = np.fft.fftshift(fftimg)
fftimglog = 100 * np.log(abs(fftimg))
centeredfftlog = 100 * np.log(abs(centeredfft))
plt.imshow(fftimglog,cmap='gray')
```

```python
plt.imshow(centeredfftlog,cmap='gray')
```

```python
filter1 =np.zeros(grayscale.shape)
filter1= cv2.rectangle(filter1,(100,100),(500,500),(255,255,255),-1)
plt.imshow(filter1,cmap='gray')


```

```python
fftfil = np.fft.fft2(filter1)
cenfftfil = np.fft.fftshift(fftfil)
logcenfftfil = 100 * np.log(abs(cenfftfil))
newimg = logcenfftfil*centeredfftlog
```

```python

filteredImg = np.abs(np.fft.ifft2(newimg))
plt.imshow(filteredImg,cmap='gray')
```
