```python

import cv2
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
print(rgb.shape)
plt.imshow(rgb)
```

```python
grayscale= cv2.cvtColor(rgb,cv2.COLOR_BGR2GRAY)
```

```python
mask = np.zeros(rgb.shape[:2])
mask[200:1000,200:800] = 255
print(mask.shape)
# mask = np.dstack((mask,mask,mask))
print(mask.shape)
plt.imshow(mask,cmap='binary')
```

```python
binary_mask = np.zeros(grayscale.shape, dtype=np.uint8)
binary_mask = cv2.circle(binary_mask, (600, 400), 300, (255,255,255), -1)
result = grayscale & binary_mask

plt.imshow(result,cmap='gray')
```
