```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
plt.imshow(rgb)
```

```python
grayscale = cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale,cmap='gray')
```

```python
print(grayscale.shape)
gra2 = np.copy(grayscale);
gra2= gra2.reshape(-1)
print(gra2.shape)
xlavel = np.arange(0,256)
ylavel = np.zeros(256)
print(xlavel.shape)
for i in range(len(gra2)):
    ylavel[gra2[i]] += 1;
# print(xlavel)
plt.bar(xlavel,ylavel,width=1)
```

```python
plt.subplot(1,1,1)
plt.title('histogram')
plt.hist(grayscale.ravel(),255,[0,256])
```
