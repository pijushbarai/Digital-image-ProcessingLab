```python

import cv2
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\tree.jpg'
rgb = plt.imread(img_path)
print(rgb.shape)
plt.imshow(rgb)
```

```python
# grayscale= cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
_,binary = cv2.threshold(rgb,127,255,cv2.THRESH_BINARY)
plt.imshow(binary,cmap='gray')
binary ^= 255
```

```python
kernel = np.ones((3,3))
w,h = binary.shape
newimage = np.zeros((w-2,h-2))
for i in range(w-2):
    for j in range(h-2):
        mat = binary[i:i+3,j:j+3]
        newimage[i,j] =  np.sum(np.multiply(mat,kernel))
        if(newimage[i,j]> 0):
            newimage[i,j] = 255
plt.imshow(newimage,cmap='binary')
```

```python
newimage2 = np.zeros((w-2,h-2))
for i in range(w-2):
    for j in range (h-2):
        mat = binary[i:i+3,j:j+3]
        val = np.sum(np.multiply(mat,kernel))
        if(val == (9*255)):
            newimage2[i,j] = 255
        else:
            newimage2[i,j] = 0
plt.imshow(newimage2,cmap='binary')
```
