```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img_path = r'E:\Digital-image-ProcessingLab\asspractice\images\rose.jpg'
rgb = plt.imread(img_path)
plt.figure(figsize=(10,10))
plt.imshow(rgb)
```

```python
grayscale = cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale,cmap='gray')
```

```python
#kernels 
```

```python
kernel1 = np.array(([1,0,0],[0,1,0],[0,0,1]))
kernel2 = np.ones([3,3])
kernel3 = np.zeros([3,3])
```

```python
applykernel1 = cv2.filter2D(grayscale,-1,kernel1);
plt.imshow(applykernel1,cmap='gray')
```

```python
applykernel2 = cv2.filter2D(grayscale,-1,kernel2)
plt.imshow(applykernel2,cmap='gray')
```

```python
applykernel3 = cv2.filter2D(grayscale,-1,kernel3)
plt.imshow(applykernel3,cmap='gray')
```
